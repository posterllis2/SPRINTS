```python
SYSTEM_PROMPT = """
Você é o ChargeOps AI, o Assistente Inteligente especialista na solução EV ChargeOps da GoodWe. Seu objetivo principal é auxiliar síndicos, zeladores e administradores de condomínios na gestão, faturamento e monitoramento da infraestrutura de recarga de veículos elétricos (EV).

DIRETRIZES DE COMPORTAMENTO:
1. Tom e Estilo: Mantenha uma postura profissional, técnica, porém prestativa e de fácil compreensão para quem gerencia o condomínio.
2. Escopo de Conhecimento: Você domina a orquestração dinâmica de potência (ChargeGrid Intelligence), faturamento e rateio por morador baseado nas normas da ANEEL, monitoramento em tempo real via protocolo OCPP e segurança elétrica predial.
3. Blindagem de Escopo: Se o usuário fizer perguntas fora do escopo de recarga de veículos elétricos ou fora das soluções GoodWe, responda cordialmente: "Lamento, mas só posso ajudar com dúvidas relacionadas à gestão de recarga de veículos elétricos e soluções EV ChargeOps da GoodWe."
4. Transição para Suporte Humano: Se a dúvida exigir manutenção física ou análise de hardware específico, direcione para o suporte técnico humano da GoodWe.
"""
```

```python
import os

try:
    from google.colab import userdata
    OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
except ImportError:
    from dotenv import load_dotenv
    load_dotenv()
    OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

from openai import OpenAI

if not OPENAI_API_KEY:
    raise ValueError("Erro: A API Key 'OPENAI_API_KEY' não foi encontrada. Configure-a antes de rodar.")

client = OpenAI(api_key=OPENAI_API_KEY)

#Inicialização do histórico de mensagens com o papel do "system"
historico_conversa = [
    {"role": "system", "content": SYSTEM_PROMPT}
]

print("=== EV ChargeOps AI (OpenAI) - Atendimento GoodWe Iniciado ===")
print("Digite 'sair' para encerrar a sessão.\n")

#Loop de interação contínua
while True:
    user_input = input("Usuário: ")
    if user_input.lower() == 'sair':
        print("Sessão encerrada.")
        break

    if not user_input.strip():
        continue

    # Adiciona a fala do usuário ao histórico
    historico_conversa.append({"role": "user", "content": user_input})

    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=historico_conversa,
            temperature=0.3 # Mantém as respostas técnicas e consistentes
        )

        #Recupera o texto da resposta
        bot_response = response.choices[0].message.content
        print(f"IA: {bot_response}\n")

        #Salva a resposta do bot no histórico para manter a memória no próximo turno
        historico_conversa.append({"role": "assistant", "content": bot_response})

    except Exception as e:
        print(f"Erro ao processar mensagem: {e}\n")
        # Remove a última mensagem do usuário do histórico se deu erro, para não quebrar a sequência lógica
        historico_conversa.pop()
```
